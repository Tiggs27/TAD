```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False
```

```python

bank  = pd.read_csv("bank_data_v1.csv",delimiter=",")
bank["pdays"]=bank["pdays"].replace([999],0)
bank = bank.drop(['months','day_of_week','duration'],axis=1)
job_values = bank["job"].unique().tolist()
str_job ="["
for i in range(len(job_values)):
    str_job += job_values[i]
    if i != len(job_values)-1:
        str_job +="/"
str_job+="]"
edu_values = bank["education"].unique().tolist()
str_edu ="["
for i in range(len(edu_values)):
    str_edu += edu_values[i]
    if i != len(edu_values)-1:
        str_edu +="/"
str_edu+="]"
marital_values = bank["marital"].unique().tolist()
str_mari ="["
for i in range(len(marital_values)):
    str_mari += marital_values[i]
    if i != len(marital_values)-1:
        str_mari +="/"
str_mari+="]"
credit_values = bank["credit"].unique().tolist()
str_credi ="["
for i in range(len(credit_values)):
    str_credi += credit_values[i]
    if i != len(credit_values)-1:
        str_credi +="/"
str_credi+="]"
housing_values = bank["housing"].unique().tolist()
str_housing ="["
for i in range(len(housing_values)):
    str_housing += housing_values[i]
    if i != len(housing_values)-1:
        str_housing +="/"
str_housing+="]"
loan_values = bank["loan"].unique().tolist()
str_loan ="["
for i in range(len(loan_values)):
    str_loan += loan_values[i]
    if i != len(loan_values)-1:
        str_loan +="/"
str_loan+="]"
poutcome_values = bank["poutcome"].unique().tolist()
str_poutcome ="["
for i in range(len(poutcome_values)):
    str_poutcome += poutcome_values[i]
    if i != len(poutcome_values)-1:
        str_poutcome +="/"
str_poutcome+="]"
```

```python
def preprocessing(bank):
    categorical_values = bank.select_dtypes(exclude = 'number')
    numerical_values=bank.select_dtypes(include='number')
    scale = StandardScaler()
    
    le = LabelEncoder()
    numerical_values_scaled = scale.fit_transform(numerical_values)
    new_cat_values = categorical_values.copy()
    columns = new_cat_values.columns

    for i in range(categorical_values.shape[1]):
        label = le.fit_transform(new_cat_values[columns[i]])
        new_cat_values.drop(columns[i],axis=1,inplace = True)
        new_cat_values[columns[i]] = label
    df_num_val = pd.DataFrame(numerical_values_scaled,columns = numerical_values.columns)
    total_columns = pd.concat([df_num_val, new_cat_values], axis=1)
    y = total_columns["outcome"]
    x = total_columns.drop(['outcome'],axis=1)
    x.to_csv("X_dt.csv")
    y.to_csv("Y_dt.csv")
    return x,y
```

```python
def get_parameters():
  print("PREDICT THE PERSON'S DECISION\n")
  
  age_input = input("Enter the age:")
  while age_input.isnumeric() != True:
    age_input = input("Enter the age:")
  age_input=int(age_input)

  job_input=input("Enter the job "+str_job)
  while job_input not in job_values:
    job_input=input("Enter the job"+str_job)

  education_input=input("Enter the education "+str_edu)
  while education_input not in edu_values:
    education_input=input("Enter the education"+str_edu)

  marital_input=input("Enter the marital status "+str_mari)
  while marital_input not in marital_values:
    marital_input=input("Enter the marital status "+str_mari)
                          
  credit_input=input("Enter the credit "+str_credi)
  while credit_input not in credit_values:
    credit_input=input("Enter the credit "+str_credi)

  housing_input=input("Enter if has a housing loan "+str_housing)
  while housing_input not in housing_values:
    housing_input=input("Enter if has a housing loan "+str_housing)

  loan_input=input("Enter if has a personal loan "+str_loan)
  while loan_input not in loan_values:
    loan_input=input("Enter if has a personal loan "+str_loan)

  poutcome_input=input("Enter the previous outcome in the previous outcome "+str_poutcome)
  while poutcome_input not in poutcome_values:
    poutcome_input=input("Enter the previous outcome in the previous outcome "+str_poutcome)


  emp_var_rate_input = input("Enter the employment variation rate:")
  while isfloat(emp_var_rate_input)!=True:
    emp_var_rate_input = input("Enter the employment variation rate:")
  emp_var_rate_input = float(emp_var_rate_input)

  cons_price_idx_input = input("Enter the consumer price index:")
  while isfloat(cons_price_idx_input) !=True:
    cons_price_idx_input = input("Enter the consumer price index:")
  cons_price_idx_input = float(cons_price_idx_input)

  cons_conf_idx_input = input("Enter the consumer confidence index:")
  while isfloat(cons_conf_idx_input) !=True:
    cons_conf_idx_input = input("Enter the consumer confidence index:")
  cons_conf_idx_input = float(cons_conf_idx_input)

  euribor3m_input = input("Enter the euribor3m:")
  while isfloat(euribor3m_input) !=True:
    euribor3m_input = input("Enter the euribor 3 months rate:")
  euribor3m_input = float(euribor3m_input)

  n_employed_input = input("Enter the number employees rate:")
  while isfloat(n_employed_input) != True:
    n_employed_input = input("Enter the number employees rate:")
  n_employed_input = float(n_employed_input)

  n_contacts_campaign_input = input("Enter number of contacts performed during this campaign and for this client:")
  while n_contacts_campaign_input.isnumeric() !=True:
    n_contacts_campaign_input=input("Enter number of contacts performed during this campaign and for this client:")
  n_contacts_campaign_input = int(n_contacts_campaign_input)

  n_contacts_previous_input = input("Enter number of contacts performed in a previous campaign for this client:")
  while n_contacts_previous_input.isnumeric() !=True:
    n_contacts_previous_input=input("Enter number of contacts performed in a previous campaign for this client:")
  n_contacts_previous_input = int(n_contacts_previous_input)

  pdays_input = input("Enter number of days that passed by after the client was last contacted from a previous campaign: ")
  while pdays_input.isnumeric() !=True:
    pdays_input=input("Enter number of days that passed by after the client was last contacted from a previous campaign: ")
  pdays_input = int(pdays_input)
  
  cat = pd.DataFrame(np.array([job_input,marital_input,education_input,credit_input,housing_input,loan_input,poutcome_input]),index= ["job","marital","education","credit","housing","loan","poutcome"])
  numb = pd.DataFrame(np.array([age_input,n_contacts_campaign_input,emp_var_rate_input,cons_price_idx_input,cons_conf_idx_input,euribor3m_input,n_employed_input,pdays_input,n_contacts_previous_input]),index = ["age","n_contacts_campaign","emp_var_rate","cons_price_idx","cons_conf_idx","euribor3m","nr_employed","pdays","n_contacts_previous"])
  return cat,numb
```

```python
from numpy.core import numeric
def values_standarization(cat,numb):
  #colocados no index 1
  categorical_values = bank.select_dtypes(exclude = 'number')
  categorical_values = categorical_values.drop(["outcome"],axis = 1)
  #display(categorical_values)
  categorical_values.loc[0] = cat.loc["job",0],cat.loc["marital",0],cat.loc["education",0],cat.loc["credit",0],cat.loc["housing",0],cat.loc["loan",0],cat.loc["poutcome",0]
  numerical_values=bank.select_dtypes(include='number')
  
  numerical_values.loc[0] = numb.loc["age",0],numb.loc["n_contacts_campaign",0],numb.loc["emp_var_rate",0],numb.loc["cons_price_idx",0],numb.loc["cons_conf_idx",0],numb.loc["euribor3m",0],numb.loc["nr_employed",0],numb.loc["pdays",0],numb.loc["n_contacts_previous",0]
  numerical_values = numerical_values.sort_index().reset_index(drop=True)


  
  #display(bank)
  #display(numerical_values)
  scale = StandardScaler()
      
  le = LabelEncoder()
  num_scaled = scale.fit_transform(numerical_values)
  
  new_cat_values = categorical_values.copy()
  
  columns = new_cat_values.columns

  for i in range(categorical_values.shape[1]):
      label = le.fit_transform(new_cat_values[columns[i]])
      new_cat_values.drop(columns[i],axis=1,inplace = True)
      new_cat_values[columns[i]] = label
  #display(new_cat_values)
  #array_input_values = [num_scaled[0,0],num_scaled[0,1],num_scaled[0,2],num_scaled[0,3],num_scaled[0,4],num_scaled[0,5],num_scaled[0,6],num_scaled[0,7],num_scaled[0,8],new_cat_values[0,0],new_cat_values[0,1],new_cat_values[0,2],new_cat_values[0,3],new_cat_values[0,4],new_cat_values[0,5],new_cat_values[0,6]]
  #array_input_values= np.concatenate((num_scaled[0,:], new_cat_values.to_numpy()[0,:]))
  #print(array_input_values)
  
  df_numb = pd.DataFrame(num_scaled,columns = numerical_values.columns)
  new_columns_val = pd.concat([df_numb, new_cat_values], axis=1)
  return new_columns_val
```

```python
def train_model(x,y):
    rf = RandomForestClassifier(bootstrap=False,max_depth= None,max_samples = None,min_samples_leaf= 10,min_samples_split= 8, n_estimators= 200)
    model_rf = rf.fit(x,y)
    return model_rf
```

```python

if __name__ == "__main__":
  
    x,y=preprocessing(bank)
    model = train_model(x,y)
    continue_var = True
    while continue_var:
      cat,numb = get_parameters()
      input_values= values_standarization(cat,numb)
      predited_values = model.predict(input_values)
      #print(predited_values[0])
      if predited_values[0] == 0:
          print("REJECT :(\n")
      else:
          print("APPROVED!!")
      continue_input = input("Do you want to continue?[yes/no]: ")
      while continue_input not in ["yes","no"]:
          continue_input = input("Do you want to continue?[yes/no]: ")
      if continue_input == "no":
          continue_var = False
      else:
          continue_var = True


```
